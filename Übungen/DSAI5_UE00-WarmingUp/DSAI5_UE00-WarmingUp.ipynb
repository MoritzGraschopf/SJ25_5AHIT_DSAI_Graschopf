```python
movie_df["has_seen"] = 0
```

```python
cols = [
    "director_facebook_likes",
    "actor_1_facebook_likes",
    "actor_2_facebook_likes",
    "actor_3_facebook_likes",
]

affected_rows_mask = movie_df[cols].isna().any(axis=1)
print("Anzahl betroffener Einträge:", int(affected_rows_mask.sum()))
movie_df[cols] = movie_df[cols].fillna(0)

movie_df["director_actor_facebook_likes"] = movie_df[cols].sum(axis=1).astype(int)

movie_df
```

```python
movie_df["is_expensive"] = movie_df["budget"] > 22000000
movie_df[["is_expensive", "budget"]]
```

```python
spam_df = pd.read_csv("spam.csv")

unnamed_cols = [c for c in spam_df.columns if c.lower().startswith("unnamed:")]
if "v2" in spam_df.columns and unnamed_cols:
    spam_df["v2"] = (
        spam_df[["v2"] + unnamed_cols]
        .astype(str)
        .replace({"nan": ""})
        .agg(" ".join, axis=1)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
    )
    spam_df = spam_df.drop(columns=unnamed_cols)
```

```python
spam_df["length"] = spam_df["v2"].str.len()
spam_df
```
